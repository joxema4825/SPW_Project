```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
```

```python
mask_path = r"C:\Users\Dell\Downloads\prediction.nii.gz"
input_flair_path = r"E:\SPW\Dataset\MICCAI_BraTS2020_ValidationData\BraTS20_Validation_002\BraTS20_Validation_002_flair.nii.gz"
input_t1_path = r"E:\SPW\Dataset\MICCAI_BraTS2020_ValidationData\BraTS20_Validation_002\BraTS20_Validation_002_t1.nii.gz"
input_t1ce_path = r"E:\SPW\Dataset\MICCAI_BraTS2020_ValidationData\BraTS20_Validation_002\BraTS20_Validation_002_t1ce.nii.gz"
input_t2_path = r"E:\SPW\Dataset\MICCAI_BraTS2020_ValidationData\BraTS20_Validation_002\BraTS20_Validation_002_t2.nii.gz"
```

```python
flair_data = nib.load(input_flair_path).get_fdata()
t1_data = nib.load(input_t1_path).get_fdata()
t1ce_data = nib.load(input_t1ce_path).get_fdata()
t2_data = nib.load(input_t2_path).get_fdata()
mask_data = nib.load(mask_path).get_fdata()
```

```python
plt.subplot(2, 4, 1)
plt.imshow(flair_data[:,:,75])
plt.axis('off')

plt.subplot(2, 4, 2)
plt.imshow(t1_data[:,:,75])
plt.axis('off')

plt.subplot(2, 4, 3)
plt.imshow(t1ce_data[:,:,75])
plt.axis('off')

plt.subplot(2, 4, 4)
plt.imshow(t2_data[:,:,75])
plt.axis('off')

plt.subplot(2, 4, 5)
plt.imshow(mask_data[:,:,75])
plt.imshow(flair_data[:,:,75], alpha=0.4, cmap='grey')
plt.axis('off')

plt.subplot(2, 4, 6)
plt.imshow(mask_data[:,:,75])
plt.imshow(t1_data[:,:,75], alpha=0.4, cmap='grey')
plt.axis('off')

plt.subplot(2, 4, 7)
plt.imshow(mask_data[:,:,75])
plt.imshow(t1ce_data[:,:,75], alpha=0.4, cmap='grey')
plt.axis('off')

plt.subplot(2, 4, 8)
plt.imshow(mask_data[:,:,75])
plt.imshow(t2_data[:,:,75], alpha=0.4, cmap='grey')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
labels = {
    0: 'Background',
    1: 'Necrosis',
    2: 'Edema',
    3: 'Tumor Core'
}
colors = ['black', 'red', 'green', 'blue']
cmap = ListedColormap(colors)

# Plot
plt.imshow(mask_data[:,:,75], cmap=cmap)
plt.imshow(flair_data[:,:,75], alpha=0.6, cmap='grey')
plt.title("Segmentation with Labels")
plt.axis('off')

# Create legend
legend_handles = [Patch(color=colors[i], label=labels[i]) for i in labels]
plt.legend(handles=legend_handles, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
