```python
import nibabel as nib
import numpy as np
```

```python
def normalize(nparray):
  nparraymean = np.mean(nparray[nparray > 0])
  nparraystd = np.std(nparray[nparray > 0])

  return (nparray - nparraymean)/nparraystd if nparraystd != 0 else nparray


def preprocessor(path_to_subfolder, i):
  print(path_to_subfolder)
  # /content/drive/MyDrive/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/BraTS20_Training_001_flair.nii.gz
  flair_path = f"{path_to_subfolder}/BraTS2021_{i:0{3}d}_flair.nii.gz"
  t1_path = f"{path_to_subfolder}/BraTS2021_{i:0{3}d}_t1.nii.gz"
  t1ce_path = f"{path_to_subfolder}/BraTS2021_{i:0{3}d}_t1ce.nii.gz"
  t2_path = f"{path_to_subfolder}/BraTS2021_{i:0{3}d}_t2.nii.gz"
  seg_path = f"{path_to_subfolder}/BraTS2021_{i:0{3}d}_seg.nii.gz"

  flair_data = nib.load(flair_path).get_fdata()
  t1_data = nib.load(t1_path).get_fdata()
  t1ce_data = nib.load(t1ce_path).get_fdata()
  t2_data = nib.load(t2_path).get_fdata()
  mask_data = nib.load(seg_path).get_fdata()
  mask_data[mask_data == 4] = 3

  flair_data = normalize(flair_data)
  t1_data = normalize(t1_data)
  t2_data = normalize(t2_data)
  t1ce_data = normalize(t1ce_data)

  for j in range(1, 155):
    slice_path_input = f"./Preprocessed_data/InputSlice_{i+295:0{3}d}_{j:0{3}d}.npz"
    slice_path_mask = f"./Preprocessed_data/MaskSlice_{i+295:0{3}d}_{j:0{3}d}.npz"

    flair_slice = flair_data[:, :, j]
    t1_slice = t1_data[:, :, j]
    t1ce_slice = t1ce_data[:, :, j]
    t2_slice = t2_data[:, :, j]

    input_slice = np.stack([flair_slice, t1_slice, t1ce_slice, t2_slice], axis=-1)
    mask_slice = mask_data[:, :, j]
    
    # np.save(slice_path_input, input_slice)
    # np.save(slice_path_mask, mask_slice)

    np.savez_compressed(slice_path_input, input=input_slice)
    np.savez_compressed(slice_path_mask, mask=mask_slice)
```

```python
path_to_dirs = 'E:/Preprocess/New_Folder/'
# /content/drive/MyDrive/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001
for i in range(63, 334):
  path_to_subdir = f"{path_to_dirs}BraTS2021_{i:0{3}d}"
  preprocessor(path_to_subdir, i)
```
